# Sub Plot

```python
plt.plot(a,b)
```

```python
plt.plot(a,c)
```

```python
plt.plot(a,d)
```

```python
#plot 1
plt.subplot(2,2,2)
plt.plot(a,d)

#plot 2
plt.subplot(2,2,3)
plt.plot(a,c)
```

```python
#plot 1
plt.subplot(2,2,1)
plt.plot(a,b)

#plot 1
plt.subplot(2,2,2)
plt.plot(a,c)
#plot 1
plt.subplot(2,2,3)
plt.plot(a,d)

#plot 2
plt.subplot(2,2,4)
plt.plot(c,a)
```

```python
#plot 1
plt.subplot(3,2,1)
plt.plot(a,b)

#plot 2
plt.subplot(3,2,2)
plt.plot(a,c)

#plot 3
plt.subplot(3,2,5)
plt.plot(a,d)

#plot 4
plt.subplot(3,2,4)
plt.plot(c,a)
```

```python
#plot 1
plt.subplot(3,2,1)
plt.plot(a,b)

#plot 2
plt.subplot(3,2,2)
plt.plot(a,c)

#plot 3
plt.subplot(3,2,3)
plt.plot(a,d)

#plot 4
plt.subplot(3,2,4)
plt.plot(c,a)

#plot 5
plt.subplot(3,2,5)
plt.plot(d,c)

#plot 6
plt.subplot(3,2,6)
plt.plot(c,b)
```

```python
#plot 1
plt.subplot(3,2,1)
plt.plot(a,b,color='r')

#plot 2
plt.subplot(3,2,2)
plt.plot(a,c,color='g')

#plot 3
plt.subplot(3,2,3)
plt.plot(a,d,color='b')

#plot 4
plt.subplot(3,2,4)
plt.plot(c,a,color='brown')

#plot 5
plt.subplot(3,2,5)
plt.plot(d,c,color='violet')

#plot 6
plt.subplot(3,2,6)
plt.plot(c,b,color='y')
```
